```python
this_policies = [i for i in this_df.columns if i in policy_cols]
this_epi = [i for i in this_df.columns if i in epi_cols]
this_id = [i for i in this_df.columns if i in id_cols]
bad_cols = [i for i in this_df.columns if i not in this_policies+this_epi+this_id]
assert len(bad_cols) == 0, bad_cols
```
